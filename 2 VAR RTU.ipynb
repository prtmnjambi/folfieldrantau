```python
print('Variabel Untuk Uji Regresi Data')
columns_to_drop = ['Tanggal', 'Jam(WIB)', 'D1', 'D2', 'D3']  # Ganti dengan nama kolom yang ingin dihapus

# Menghapus kolom yang tidak diperlukan
x = data_Latih.drop(columns=columns_to_drop)

# Mengambil kolom 'D1', 'D2', dan 'D3' sebagai output
out_cols = data_Latih[['D1', 'D2', 'D3']]
y = out_cols
```

```python
import statsmodels.api as sm

# Menambahkan konstanta ke variabel independen
X = sm.add_constant(x)

# Membuat model OLS untuk setiap variabel dependen secara terpisah
results = []
for col in y.columns:
    model = sm.OLS(y[col], X).fit()
    results.append((col, model))

# Mencetak ringkasan setiap model
for col, model in results:
    print(f"Model for dependent variable: {col}")
    print(model.summary())
    print("\n")

# Menampilkan R-squared untuk setiap model
for col, model in results:
    print(f"R-squared for {col}:", model.rsquared)
```

```python
import statsmodels.api as sm

# Menambahkan konstanta ke variabel independen
X1 = sm.add_constant(x)  # Variabel independen untuk Y1
X2 = sm.add_constant(x)  # Variabel independen untuk Y2
X3 = sm.add_constant(x)  # Variabel independen untuk Y3

# Ganti nama kolom "Titik_1_PSI" hingga "Titik_9_PSI" menjadi "x1" hingga "x9"
X1.columns = ["const"] + [f"x{i+1}" for i in range(X1.shape[1] - 1)]
X2.columns = ["const"] + [f"x{i+1}" for i in range(X2.shape[1] - 1)]
X3.columns = ["const"] + [f"x{i+1}" for i in range(X3.shape[1] - 1)]

# Menyusun variabel dependen
Y = {"Y1": y["D1"], "Y2": y["D2"], "Y3": y["D3"]}

# Membuat model OLS untuk setiap variabel dependen dengan variabel independennya masing-masing
results = []
for label, X in zip(Y.keys(), [X1, X2, X3]):
    model = sm.OLS(Y[label], X).fit()
    results.append((label, model))

# Menampilkan persamaan regresi untuk setiap model
for label, model in results:
    # Mengambil koefisien model
    params = model.params
    # Membuat persamaan regresi
    equation = f"{label} = {params[0]:.4f}"
    for i, name in enumerate(params.index[1:], start=1):
        equation += f" + ({params[name]:.4f}) * {name}"
    print(f"Persamaan regresi untuk {label}:\n{equation}\n")

# Mencetak ringkasan setiap model
for label, model in results:
    print(f"Model for dependent variable: {label}")
    print(model.summary())
    print("\n")

# Menampilkan R-squared untuk setiap model
for label, model in results:
    print(f"R-squared for {label}:", model.rsquared)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Data dari tabel
data = {
    "Location": [
        "PPP RTU 0 km", "8.3 km (SP 12)", "15.9 km (Sei. Liput)",
        "24.9 km (Karangjadi)", "30.8 km (Karetia)", "38.9 km (Sisirah)",
        "44.5 km (Kp. Lama)", "56.5 km (Pipa 8)", "63 km (Tml. PSU)"
    ],
    "Delta_Pressure_1": [1.46, 2.87, 8.5, 3.55, 3.14, 2.66, 1.5, 0.66, 0.0],
    "Delta_Pressure_2": [5.62, 3.13, 2.9, 1.01, 0.0, 0.0, 0.8, 0.94, 0.44]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Membuat dataset input untuk model
x_test = np.array([0, 8.3, 15.9, 24.9, 30.8, 38.9, 44.5, 56.5, 63.0]).reshape(-1, 1)
y_test = np.array(df["Delta_Pressure_1"])

# Model Decision Tree
Dc = DecisionTreeRegressor(max_depth=6, random_state=30)
Dc.fit(x_test, y_test)

# Prediksi delta tekanan di sepanjang jalur pipa
Distance_of_Pipeline = np.linspace(0, 63.4, 200).reshape(-1, 1)
predictions_now = Dc.predict(Distance_of_Pipeline)

# Mencari lokasi dengan delta tekanan maksimum
max_idx = np.argmax(predictions_now)
max_km = Distance_of_Pipeline[max_idx][0]
max_delta = predictions_now[max_idx]

# Menampilkan hasil
print(f"Max Delta terjadi di KM: {max_km:.2f} dengan nilai delta: {max_delta:.2f}")

# Membuat grafik prediksi
plt.figure(figsize=(12, 7))

# Plot data asli dan prediksi
plt.plot(x_test.flatten(), y_test, 'o', label='Delta Pressure (Data Asli)', markersize=8)
plt.plot(Distance_of_Pipeline, predictions_now, '-', label='Prediksi (Decision Tree)', linewidth=2)

# Menandai posisi delta maksimum
plt.axvline(x=max_km, color='red', linestyle='--', label=f'Max Delta (KM {max_km:.2f}, {max_delta:.2f})')

# Menambahkan teks untuk setiap titik data asli
for km, delta in zip(x_test.flatten(), y_test):
    plt.text(km, delta + 0.5, f"KM: {km}, Delta: {delta:.2f}", fontsize=10, ha='center')

# Menambahkan teks untuk prediksi maksimum
plt.text(max_km, max_delta + 0.5, f"Max Delta Prediksi\n(KM {max_km:.2f}, {max_delta:.2f})", fontsize=12, color='red')

# Mengatur label dan judul
plt.xlabel('Jarak (KM)', fontsize=12)
plt.ylabel('Delta Tekanan', fontsize=12)
plt.title('Prediksi Delta Tekanan Pipa', fontsize=14)
plt.legend()
plt.grid(True)

# Menampilkan grafik
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Data tekanan normal dan drop di berbagai lokasi
data = {
    "Location": [
        "PPP RTU 0 km", "8.3 km (SP 12)", "15.9 km (Sei. Liput)",
        "24.9 km (Karangjadi)", "30.8 km (Karetia)", "38.9 km (Sisirah)",
        "44.5 km (Kp. Lama)", "56.5 km (Pipa 8)", "63 km (Tml. PSU)"
    ],
    "Normal": [493.98, 446.91, 410.77, 308.72, 206.58, 192.50, 133.86, 70.90, 0.93],
    "Drop": [492.04, 444.48, 392.77, 307.03, 205.49, 192.05, 133.30, 70.82, 0.86

]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Menghitung delta tekanan (perbedaan tekanan normal dan drop)
df['Delta'] = df['Normal'] - df['Drop']

# Mengambil data jarak (KM) dan delta tekanan
locations_km = [0, 8.3, 15.9, 24.9, 30.8, 38.9, 44.5, 56.5, 63.0]
delta_values = df['Delta'].values

# Melatih model Random Forest Regressor
X_train = np.array(locations_km).reshape(-1, 1)
y_train = delta_values
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Memprediksi delta tekanan di sepanjang jalur pipa (0 hingga 70 km)
Distance_of_Pipeline = np.linspace(0, 63.4, 200)
predictions_now = model.predict(Distance_of_Pipeline.reshape(-1, 1))

# Mencari lokasi dengan delta tekanan maksimum
max_idx = np.argmax(predictions_now)
max_km = Distance_of_Pipeline[max_idx]
max_delta = predictions_now[max_idx]

# Menampilkan hasil
print(f"Max Delta terjadi di KM: {max_km:.2f} dengan nilai delta: {max_delta:.2f}")

# Membuat grafik prediksi
plt.figure(figsize=(12, 7))
plt.plot(locations_km, delta_values, 'o', label='Data Asli', markersize=8)
plt.plot(Distance_of_Pipeline, predictions_now, '-', label='Prediksi (Random Forest)', linewidth=2)
plt.axvline(x=max_km, color='red', linestyle='--', label=f'Max Delta (KM {max_km:.2f}, {max_delta:.2f})')

# Menambahkan teks untuk setiap titik data asli dan prediksi
for km, delta in zip(locations_km, delta_values):
    plt.text(km, delta + 1, f"KM: {km}, Delta: {delta:.2f}", fontsize=10, ha='center')

plt.text(max_km - 12, max_delta + 0.5, f"Max Delta Prediksi\n(KM {max_km:.2f}, {max_delta:.2f})", fontsize=12, color='red')

# Mengatur label dan judul
plt.xlabel('Jarak (KM)', fontsize=12)
plt.ylabel('Delta Tekanan', fontsize=12)
plt.title('Prediksi Delta Tekanan Pipa', fontsize=14)
plt.legend()
plt.grid(True)

# Menampilkan grafik
plt.show()
```
